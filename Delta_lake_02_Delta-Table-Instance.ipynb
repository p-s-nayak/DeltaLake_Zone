##### i. To make soft link(replica) to actual delta table.
##### ii.Usage of Delta Table Instance: To perform DML operations using Pyspark language.

####Syntax: 
##### . Using Path - DeltaTable.forPath(spark,”/path/to/table”)
##### . Using TableName - DeltaTable.forName(spark,”Table_Name”)

#### 1- Using Path

```
from delta.tables import *
DeltaTable.create(spark)\
    .tableName("employee_demo_instance") \
    .addColumn("emp_id","INT") \
    .addColumn("emp_name","STRING") \
    .addColumn("gender","STRING") \
    .addColumn("salary","INT") \
    .addColumn("Dept","String") \
    .property("Description","Table for employee") \
    .location("/FileStore/tables/delta/createtable") \
    .execute()
```

```
%sql
INSERT INTO employee_demo_instance VALUES(100,"Stephen","M",10000,"IT");
INSERT INTO employee_demo_instance VALUES(200,"Sare","F",13000,"IT");
INSERT INTO employee_demo_instance VALUES(300,"Subha","M",9000,"Admin");
INSERT INTO employee_demo_instance VALUES(400,"zara","M",80000,"Sales");
```

```
%sql
select * from employee_demo_instance
```

```
deltaInstance1 = DeltaTable.forPath(spark,"/FileStore/tables/delta/createtable")
```

```
display(deltaInstance1.toDF())
```

```
%sql 
delete from employee_demo_instance where emp_id = 100
```

```
%sql
select * from employee_demo_instance
```

```
deltaInstance1.delete("emp_id = 200")
```

```
display(deltaInstance1.toDF())
```

##2- Using TableName. :forName

```
deltaInstance2=DeltaTable.forName(spark,"employee_demo_instance")
```

```
display(deltaInstance2.toDF())
```

```
%sql
DESCRIBE HISTORY employee_demo_instance
```

```
display(deltaInstance2.history())
```
